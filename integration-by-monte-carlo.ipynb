```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the integrand function
def integrand(x):
    return np.exp(x) * np.cos(x)
```

```python
def monte_carlo_simul(n):
    x_values = np.random.uniform(-np.pi/2, np.pi/2, size=n)
    y_values = np.random.uniform(0, np.exp(np.pi/2), size=n)

    # Count the number of hits
    n_hits = 0
    for x, y in zip(x_values, y_values):
        if y <= integrand(x):
            n_hits += 1

    # Estimate the area under the curve
    area_estimate = n_hits / n * np.exp(np.pi/2) * np.pi
    return area_estimate
```

# (A)

```python
total_points = 100000
print(monte_carlo_simul(total_points))
```

# (B)

```python
areas = []
nlist = [n for n in range(1, total_points+1, 1000)]

for n in nlist:
    areas.append(monte_carlo_simul(n))
    
plt.plot(nlist,areas)
plt.xlabel("N")
plt.ylabel("Integration")
plt.show()
```
